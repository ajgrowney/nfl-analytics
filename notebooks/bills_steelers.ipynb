```python
import matplotlib.pyplot as plt
import pandas as pd

DATA_ROOT = '../data/'
```

### Bills vs the Run 

```python
bills_def = pd.read_csv(f'{DATA_ROOT}/bills_rush_def.csv')
steelers_off = pd.read_csv(f'{DATA_ROOT}/steelers_rush_off.csv')
```

```python
bills_def_vs_rbs = bills_def.loc[bills_def['DepthChart'].str.contains('RB')]
steeler_rbs = steelers_off.loc[steelers_off['DepthChart'].str.contains('RB')]
```

```python
fig, axs = plt.subplots(1,2, figsize=(30, 10))
axs = axs.flatten()
axs[0].scatter(bills_def_vs_rbs['Week'], bills_def_vs_rbs['Yds'])
axs[1].scatter(steeler_rbs['Week'], steeler_rbs['Yds'])
for k,v in bills_def_vs_rbs.iterrows():
    axs[0].annotate(v['Player'], (v['Week'], v['Yds']))
for k,v in steeler_rbs.iterrows():
    axs[1].annotate(v['Player'], (v['Week'], v['Yds']))
plt.show()
```
